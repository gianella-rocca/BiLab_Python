```python
rec_1 = pd.read_spss( r"../../_data/endes/2019/REC0111.sav" )
rec_2 = pd.read_spss( r"../../_data/endes/2019/RE223132.sav" )
rec_3 = pd.read_spss( r"../../_data/endes/2019/RE516171.sav" )
```

```python
# Define the file paths
file_paths = [
    '../../_data/endes/2019/REC0111.sav',
    '../../_data/endes/2019/RE223132.sav',
    '../../_data/endes/2019/RE516171.sav'
]

# Define names for imported files and labels
file_names = ['rec_1', 'rec_2', 'rec_3']
var_labels = ['var_labels1', 'var_labels2', 'var_labels3']
value_labels = ['value_labels1', 'value_labels2', 'value_labels3']

# Initialize dictionaries to store dataframes and labels
dataframes = {}
variable_labels = {}
value_labels_dict = {}

# Loop through the files
for i, file_path in enumerate(file_paths):
    # Read the SPSS file
    df, meta = pyreadstat.read_sav(file_path)
    
    # Store the dataframe
    dataframes[file_names[i]] = df
    
    # Store variable labels
    variable_labels[var_labels[i]] = meta.column_labels
    
    # Store value labels
    value_labels_dict[value_labels[i]] = meta.variable_value_labels

# Access the imported data and labels
rec_1 = dataframes['rec_1']
var_labels1 = variable_labels['var_labels1']
value_labels1 = value_labels_dict['value_labels1']

rec_2 = dataframes['rec_2']
var_labels2 = variable_labels['var_labels2']
value_labels2 = value_labels_dict['value_labels2']

rec_3 = dataframes['rec_3']
var_labels3 = variable_labels['var_labels3']
value_labels3 = value_labels_dict['value_labels3']
```

# Exercise 6

```python
# Create the all data dictionary initializing all nested key values equal to zero
all_data = {'year_2019':{}, 'year_2018':{}, 'year_2017':{}, 'year_2016':{}, 'year_2015':{}}
years = [2019, 2018, 2017, 2016, 2015]
all_data
```

```python
data = ['rec_1', 'rec_2', 'rec_3']
var_labels = ['var_labels1', 'var_labels2', 'var_labels3']
value_labels = ['value_labels1', 'value_labels2', 'value_labels3']
```

```python
for x in years:
    all_file_paths = []
    all_file_paths.append(fr"../../_data/endes/{x}/REC0111.sav")
    all_file_paths.append(fr"../../_data/endes/{x}/RE223132.sav") 
    all_file_paths.append(fr"../../_data/endes/{x}/RE516171.sav")
    
    all_data[f'year_{x}']['data'] = {}  # Initialize nested dictionary for data
    all_data[f'year_{x}']['var_labels'] = {}  # Initialize nested dictionary for variable labels
    all_data[f'year_{x}']['value_labels'] = {}  # Initialize nested dictionary for value labels
    
    for i, file_path in enumerate(all_file_paths):
        
        # Read the SPSS file
        df, meta = pyreadstat.read_sav(file_path)
        df['Year'] = x
    
        # Store the dataframe
        all_data[f'year_{x}']['data'][data[i]] = df
    
        # Store variable labels
        all_data[f'year_{x}']['var_labels'][var_labels[i]] = meta.column_labels
    
        # Store value labels
        all_data[f'year_{x}']['value_labels'][value_labels[i]] = meta.variable_value_labels
```

```python
# Initialize dictionaries to store all var_labels and value_labels
all_var_labels = {}
all_value_labels = {}

# Iterate over the years in all_data
for year_key in all_data:
    year = int(year_key.split('_')[1])  # Extract the year from the key
    all_var_labels[year] = all_data[year_key]["var_labels"]
    all_value_labels[year] = all_data[year_key]["value_labels"]
```
