```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Example Usage:
# Create an instance of Circle
my_circle = Circle(radius=5.0)

# Print the area and perimeter of the circle
print(f"Area of the circle: {my_circle.area()}")
print(f"Perimeter of the circle: {my_circle.perimeter()}")
```

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal not processed.")

# Example Usage:
# Create an instance of BankAccount
my_account = BankAccount(account_number="123456")

# Deposit money
my_account.deposit(1000)

# Withdraw money
my_account.withdraw(500)

# Print the final balance
print(f"Final balance: ${my_account.balance}")
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def set_details(self, name, age):
        self.name = name
        self.age = age

    def get_details(self):
        return f"Student Name: {self.name}, Age: {self.age}"

    def is_adult(self):
        return self.age >= 18

# Example Usage:
# Create a Student instance
student1 = Student(name="John Doe", age=20)

# Get details
print(student1.get_details())

# Check if the student is an adult
if student1.is_adult():
    print("This student is an adult.")
else:
    print("This student is not an adult.")

# Set new details for the student
student1.set_details(name="Jane Doe", age=17)

# Get updated details
print(student1.get_details())

# Check again if the student is an adult
if student1.is_adult():
    print("This student is an adult.")
else:
    print("This student is not an adult.")
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Example Usage:
# Create a Rectangle instance
my_rectangle = Rectangle(length=5.0, width=3.0)

# Print the area and perimeter of the rectangle
print(f"Area of the rectangle: {my_rectangle.area()}")
print(f"Perimeter of the rectangle: {my_rectangle.perimeter()}")
```
